# **Step 2(df2):**
ambil url https://www.mediailmupengetahuan.com/2020/11/34-kode-provinsi-di-indonesia.html untuk mendapatkan data "provinsi" dan "kode wilayah", lalu disimpan di df2

```python
from bs4 import BeautifulSoup as bs
import requests
import re

url = 'https://www.mediailmupengetahuan.com/2020/11/34-kode-provinsi-di-indonesia.html'
response = requests.get(url)

# Parse the response content using BeautifulSoup
html_page = bs(response.content, 'html.parser')

# Find the list items containing city names and numbers using a regular expression
pattern = re.compile(r'<li>([^:]+): (\d+)</li>')
matches = pattern.findall(str(html_page))

# dictionary to store data
city_data = {}

# Iterate through the matches and store the data in the dictionary
for match in matches:
    city_name, city_number = match
    city_data[city_name.strip()] = int(city_number)

# Print the extracted data
for city, number in city_data.items():
    print(f'{city}: {number}')
```

# **Step 3(df3):**
 ambil df1 dan df2 untuk dijoin dengan key join provinsi untuk mendapatkan data "provinsi", "jumlah desa", dan "kode daerah", lalu menyimpan di df3

```python
from fuzzywuzzy import fuzz, process

name_mapping = {}

# similarity threshold
threshold = 50

for name1 in df1['Provinsi']:
    best_match = process.extractOne(name1, df2['Provinsi'])

    if best_match[1] >= threshold:
        name_mapping[name1] = best_match[0]


merged_df = df1.copy()
merged_df['Matching_Provinsi'] = merged_df['Provinsi'].map(name_mapping)


final_result = merged_df.merge(df2, left_on='Matching_Provinsi', right_on='Provinsi', how='inner')


final_result = final_result.drop(columns='Matching_Provinsi')
df3 = final_result.drop(columns='Provinsi_y')
df3
print(df3)

df3.to_excel('output_df3.xlsx', index=False)
```

# **step 4(df4):**
ambil data excel data_seki.xlsx dan dibuat menjadi table yang memiliki data kode provinsi, provinsi, jumlah desa, casa size dan lending volume, lalu disimpan di df4

```python
import pandas as pd

df = pd.read_excel(open('data_seki.xlsx', 'rb'),sheet_name='data seki', skiprows=1)
df

df['Casa Size'] = (df['seki_1_19'] + df['seki_1_20']).round(0).astype(int)
df['Lending Volume'] = (df['seki_1_12'] + df['seki_1_13']).round(0).astype(int)
df[['Bulan', 'Tahun']] = df['Posisi'].str.split("'", expand=True)

month_mapping = {
    'Jan': 'Januari',
    'Feb': 'Februari',
    'Mar': 'Maret',
    'Apr': 'April',
}

month_mapping_2 = {
    'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04',
    'May': '05', 'Jun': '06', 'Jul': '07', 'Aug': '08',
    'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
}

#Output of the column
df['Bulan'] = df['Bulan'].map(month_mapping)
df['provinsi'] = df['Provinsi']
df['Tahun'] = "20" + df['Tahun']
df['MMYYYY'] =  df['Bulan'].str[:3].map(month_mapping_2) + df['Tahun']

#Rearrange the columns
df4 = df[['provinsi', 'Casa Size', 'Lending Volume', 'Bulan', 'Tahun', 'MMYYYY']]
df4
print(df4)

df4.to_excel('output_df4.xlsx', index=False)
```

# **step 5(df_final):**

join df3 dan df4 untuk mendapatkan final output, lalu disimpan di df_final dan excel output : final_ouput.xlsx

```python
from fuzzywuzzy import fuzz, process

name_mapping = {}

# Set a similarity threshold
threshold = 50

for name1 in df3['Provinsi_x']:
    if name1 == 'Kalimantan Barat':
        name_mapping[name1] = '1'
    else:
        best_match = process.extractOne(name1, df4['provinsi'].str.strip())
        if best_match[1] >= threshold:
            name_mapping[name1] = best_match[0]

# dictionary to store data
city_data = []
city_row = []


merged_df = df3.copy()
merged_df['Matching_Provinsi'] = merged_df['Provinsi_x'].map(name_mapping)
final_result = merged_df.merge(df4, left_on='Matching_Provinsi', right_on='provinsi', how='inner')
#final_result = df3.merge(df4, left_on='Provinsi_x', right_on='provinsi', how='left')

# Drop table that don't need
final_result = final_result.drop(columns='Matching_Provinsi')
df_final = final_result.drop(columns='Provinsi_x')

# Reorder columns
df_final = df_final[['Code_Daerah', 'provinsi', 'Jumlah Desa', 'Casa Size', 'Lending Volume', 'Bulan', 'Tahun', 'MMYYYY']]

# Rename columns to match df4's columns
df_final = df_final.rename(columns={'Code_Daerah': 'Kode Provinsi', 'provinsi': 'Provinsi', 'MMYYYY': 'Posisi'})

# Remove numbers and characters from the 'Provinsi' column
df_final['Provinsi'] = df_final['Provinsi'].str.replace(r'[^a-zA-Z\s]', '', regex=True)

# Order table based on posisi and provinsi
df_final = df_final.sort_values(by=['Posisi', 'Provinsi'], ascending=[True, True])

# Reset the index after sorting
df_final = df_final.reset_index(drop=True)

# Print the sorted DataFrame
df_final
print(df_final)

df_final.to_excel('final_ouput.xlsx', index=False)
```
